```python
# ==========================================
# Assignment 1: PCA for Dimensionality Reduction
# Dataset: Breast Cancer Dataset
# ==========================================

# ------------------------------------------
# 1. Import Libraries
# ------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ------------------------------------------
# 2. Load Dataset
# ------------------------------------------

cancer = load_breast_cancer()

# Create DataFrame
df = pd.DataFrame(
    cancer.data,
    columns=cancer.feature_names
)

# Add target column
df['Target'] = cancer.target

print("Dataset Shape:", df.shape)

# Show first 5 rows
print("\nFirst 5 Rows:\n")
print(df.head())

# ------------------------------------------
# 3. Check Missing Values
# ------------------------------------------

print("\nMissing Values:\n")
print(df.isnull().sum())

# ------------------------------------------
# 4. Separate Features and Target
# ------------------------------------------

X = df.drop('Target', axis=1)

y = df['Target']

# ------------------------------------------
# 5. Before Standardization
# ------------------------------------------

print("\nBefore Standardization:\n")

print(X.describe())

# ------------------------------------------
# 6. Standardization
# ------------------------------------------

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

# ------------------------------------------
# 7. After Standardization
# ------------------------------------------

print("\nAfter Standardization:\n")

print(X_scaled_df.describe())

# ------------------------------------------
# 8. Apply PCA
# ------------------------------------------

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print("\nOriginal Shape:", X.shape)

print("Reduced Shape:", X_pca.shape)

# Explained variance
print("\nExplained Variance Ratio:\n")

print(pca.explained_variance_ratio_)

# ------------------------------------------
# 9. Scree Plot
# ------------------------------------------

pca_full = PCA()

pca_full.fit(X_scaled)

plt.figure(figsize=(7,5))

plt.plot(
    range(1, len(pca_full.explained_variance_ratio_) + 1),
    pca_full.explained_variance_ratio_,
    marker='o'
)

plt.xlabel("Principal Components")

plt.ylabel("Explained Variance Ratio")

plt.title("Scree Plot")

plt.grid(True)

plt.show()

# ------------------------------------------
# 10. PCA 2D Visualization
# ------------------------------------------

plt.figure(figsize=(7,5))

plt.scatter(
    X_pca[y == 0, 0],
    X_pca[y == 0, 1],
    label='Malignant',
    alpha=0.7
)

plt.scatter(
    X_pca[y == 1, 0],
    X_pca[y == 1, 1],
    label='Benign',
    alpha=0.7
)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA Visualization - Breast Cancer Dataset")

plt.legend()

plt.show()
```

```python
# ==========================================
# Assignment 1: PCA for Dimensionality Reduction
# Dataset: Wine Dataset
# ==========================================

# ------------------------------------------
# 1. Import Libraries
# ------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ------------------------------------------
# 2. Load Dataset
# ------------------------------------------

wine = load_wine()

# Create DataFrame
df = pd.DataFrame(
    wine.data,
    columns=wine.feature_names
)

# Add target column
df['Target'] = wine.target

print("Dataset Shape:", df.shape)

# Show first 5 rows
print("\nFirst 5 Rows:\n")
print(df.head())

# ------------------------------------------
# 3. Check Missing Values
# ------------------------------------------

print("\nMissing Values:\n")
print(df.isnull().sum())

# ------------------------------------------
# 4. Separate Features and Target
# ------------------------------------------

X = df.drop('Target', axis=1)

y = df['Target']

# ------------------------------------------
# 5. Before Standardization
# ------------------------------------------

print("\nBefore Standardization:\n")

print(X.describe())

# ------------------------------------------
# 6. Standardization
# ------------------------------------------

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# Convert to DataFrame
X_scaled_df = pd.DataFrame(
    X_scaled,
    columns=X.columns
)

# ------------------------------------------
# 7. After Standardization
# ------------------------------------------

print("\nAfter Standardization:\n")

print(X_scaled_df.describe())

# ------------------------------------------
# 8. Apply PCA
# ------------------------------------------

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print("\nOriginal Shape:", X.shape)

print("Reduced Shape:", X_pca.shape)

print("\nExplained Variance Ratio:\n")

print(pca.explained_variance_ratio_)

# ------------------------------------------
# 9. Scree Plot
# ------------------------------------------

pca_full = PCA()

pca_full.fit(X_scaled)

plt.figure(figsize=(7,5))

plt.plot(
    range(1, len(pca_full.explained_variance_ratio_) + 1),
    pca_full.explained_variance_ratio_,
    marker='o'
)

plt.xlabel("Principal Components")

plt.ylabel("Explained Variance Ratio")

plt.title("Scree Plot - Wine Dataset")

plt.grid(True)

plt.show()

# ------------------------------------------
# 10. PCA 2D Visualization
# ------------------------------------------

plt.figure(figsize=(7,5))

colors = ['red', 'green', 'blue']

labels = wine.target_names

for i, color, label in zip([0,1,2], colors, labels):

    plt.scatter(
        X_pca[y == i, 0],
        X_pca[y == i, 1],
        color=color,
        label=label,
        alpha=0.7
    )

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("PCA Visualization - Wine Dataset")

plt.legend()

plt.show()
```
